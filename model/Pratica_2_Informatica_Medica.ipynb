De la literatura concluimos que Las fechas suelen ser redundantes en el contexto de predicciones basadas en períodos de tiempo relativos.
En muchos casos, la información temporal ya está representada por variables derivadas como "Días (Camara - F3)" o "Minutos (Triaje - F3)", lo que hace que las fechas exactas sean innecesarias.

Las fechas exactas, como el día específico de una consulta o el momento de una operación, generalmente no tienen un impacto directo en la predicción de si un paciente será hospitalizado o dado de alta.

En lugar de usar fechas absolutas, los modelos de machine learning se benefician más de las duraciones y diferencias de tiempo que proporcionan una métrica más directa del estado del paciente y el flujo del proceso médico.

Incluir fechas como variables en un modelo puede complicar el preprocesamiento y no aportar valor adicional.
Las fechas requieren transformación adicional (por ejemplo, extraer día de la semana, mes, etc.), lo cual puede no ser necesario si ya se dispone de duraciones y tiempos relativos.

Los códigos de diagnóstico ya contienen toda la información necesaria sobre la naturaleza de la enfermedad o condición del paciente.
Mantener tanto los códigos como los nombres de los diagnósticos introduce redundancia, lo que puede complicar el modelo sin aportar información adicional.

Los códigos de diagnóstico son más consistentes y uniformes para el modelado, ya que son datos estructurados y categorizados de manera estándar.
Los nombres de los diagnósticos pueden variar en longitud y formato, lo que puede introducir ruido en los datos.

Los códigos de diagnóstico pueden ser fácilmente codificados (one-hot encoding) o utilizados directamente en el modelo, mientras que los nombres de los diagnósticos podrían requerir técnicas de procesamiento de texto más complejas.

```python
# Se hace la importación de la librería
from sklearn.impute import KNNImputer
```

```python
#Se eliminan los valores nulos de las variables numericas
imp = KNNImputer(n_neighbors=3 , weights='uniform')
X = df_numericas
X = imp.fit_transform(X)
df_numericas_imputer = pd.DataFrame(X, columns=df_numericas.columns)
df_numericas_imputer
```

```python
null_counts = df_numericas_imputer.isnull().sum()
for columna in df_numericas_imputer.columns:
    print(f"{columna}: {null_counts[columna]}")
```

```python
# Seleccionar columnas de tipo object
df_objetos = df_drop.select_dtypes(include=['object','bool'])
null_counts = df_objetos.isnull().sum()
for columna in df_objetos.columns:
    print(f"{columna}: {null_counts[columna]}")
```

```python
#Obtener la salida
df_salida = pd.get_dummies(df_objetos['Servicio_Admite'], prefix_sep='_')
#df_salida['HOSPITALIZACIÓN']
df_objetos = df_objetos.drop(columns=['Servicio_Admite'])
```

```python
#Colocar los valores correspondientes a Dx_Principal_Capitulo_Cod, si el valor es nulo
df_objetos['Dx_Principal_Capitulo_Cod'] = None

for i, elem in enumerate(df_objetos['Dx_Principal_Cod']):
    if elem[-1].isdigit():
        df_objetos.at[i, 'Dx_Principal_Capitulo_Cod'] = None
    else:
        df_objetos.at[i, 'Dx_Principal_Capitulo_Cod'] = elem
df_objetos['Dx_Principal_Capitulo_Cod'] = df_objetos['Dx_Principal_Cod'].apply(lambda x: x[:-1] if x[-1].isdigit() else x)
```

```python
#Colocar los valores correspondientes a Dx_Principal_Cod.1, si el valor es nulo o el Dx_Principal_Cod si no existe en la lista de Dx_Principal_Cod.1 (son 7 A90X,M45X, M725,R501)
mapping = df_objetos.dropna(subset=['Dx_Principal_Cod.1']).set_index('Dx_Principal_Cod')['Dx_Principal_Cod.1'].to_dict()

df_objetos['Dx_Principal_Cod.1'] = df_objetos.apply(
    lambda row: mapping.get(row['Dx_Principal_Cod'], row['Dx_Principal_Cod']) if pd.isnull(row['Dx_Principal_Cod.1']) else row['Dx_Principal_Cod.1'],
    axis=1
)
```

```python
#Si no tiene diagnostico relacionado colocar 0
df_objetos['Dx_Relacionado1_Cod'] = df_objetos['Dx_Relacionado1_Cod'].fillna(0)
df_objetos['Dx_Relacionado2_Cod_'] = df_objetos['Dx_Relacionado2_Cod_'].fillna(0)
df_objetos['Dx_Relacionado3_Cod'] = df_objetos['Dx_Relacionado3_Cod'].fillna(0)
```

```python
# Codificación ordinal, hay un orden del triaje
orden = ['TRIAJE 5', 'TRIAJE 4', 'TRIAJE 3', 'TRIAJE 2', 'TRIAJE 1','NO APLICA']
df_objetos2= df_objetos['Clasificacion_1_Triaje_Enfermeria_'].map({cat: idx for idx, cat in enumerate(orden)})
df_objetos = df_objetos.drop(columns=['Clasificacion_1_Triaje_Enfermeria_'])
```

```python
#Imputacion de las variables que tienen valores nulos
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

cols_to_impute = ['Servicio_Anterior','Ubicacion__Urgencias', 'Modalidad_Atencion',
                  'Tiempo_Duracion_Diligenciamiento_F3']

df_objetos[cols_to_impute] = imp.fit_transform(df_objetos[cols_to_impute])
```

```python
null_counts = df_objetos.isnull().sum()
for columna in df_objetos.columns:
    print(f"{columna}: {null_counts[columna]}")
```

```python
# Codificar las columnas categóricas utilizando One-Hot Encoding
df_objetos = pd.get_dummies(df_objetos)

num_columns = len(df_objetos.columns)
random_integers = np.random.randint(1, 100, num_columns)

for i, column in enumerate(df_objetos.columns):
    df_objetos[column] *= random_integers[i]
```

### **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split , KFold
```

####  Características o Variables de entrada

```python
df_concat = pd.concat([df_numericas_imputer, df_objetos, df_objetos2], axis=1)
X = df_concat.iloc[:,:]
X.head(5)
```

####  Etiquetas, Clases o Salida

```python
y = df_salida.iloc[:,4]
y.head(5)
```

```python
#Separación de muestras para train y test
X_train , X_test , y_train , y_test = train_test_split(X, y)

print("Train: ",X_train.shape , y_train.shape)
print("Test: ",X_test.shape , y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
       'n_estimators': [10, 50, 70],
       'max_depth': [None, 10, 20, 30],
              }
dt = RandomForestClassifier(random_state=123 , n_jobs=-1)
clf = GridSearchCV(dt, parameters , cv=5 , verbose=1)
clf.fit(X_train, y_train)
clf.best_estimator_
```

```python
clf.best_params_
```

```python
#Entrenar el modelo de Random Forest
best_model = RandomForestClassifier( n_estimators=50, max_depth=None, n_jobs=-1, random_state=123
                            ## Usar los mejores parámetros encontradas en el gridseach anterior.
                            )
best_model.fit(X_train , y_train)
best_model.score(X_test , y_test)
```

```python
# Hacer predicciones con el modelo entrenado
preds = best_model.predict(X_test)
```

```python
#Evaluar el modelo
acc = accuracy_score(y_test , preds)
acc
```
